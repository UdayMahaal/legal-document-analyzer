```python
# CELL 5A — Load spaCy model

import spacy

nlp = spacy.load("en_core_web_sm")
print("✅ spaCy model loaded!")
```

```python
# CELL 5B — Extract Legal Entities

def extract_legal_entities(text):
    """
    Finds important legal entities from contract text.
    Processes in chunks because spaCy has a size limit.
    """
    all_entities = {}
    chunk_size = 100000

    for i in range(0, len(text), chunk_size):
        chunk = text[i:i+chunk_size]
        doc = nlp(chunk)

        for ent in doc.ents:
            label = ent.label_
            value = ent.text.strip()

            # Only keep legally relevant entity types
            if label in ["PERSON", "ORG", "DATE", "MONEY",
                         "GPE", "LAW", "PERCENT", "CARDINAL"]:
                if label not in all_entities:
                    all_entities[label] = set()
                # Clean up: ignore very short or very long values
                if 2 < len(value) < 100:
                    all_entities[label].add(value)

    # Convert sets to sorted lists
    return {k: sorted(list(v)) for k, v in all_entities.items()}


def find_obligation_words(text):
    """
    Finds legal obligation keywords and counts them.
    These words tell us what parties MUST or MUST NOT do.
    """
    obligations = [
        "shall", "must", "is required to",
        "agrees to", "undertakes to",
        "will not", "shall not", "may not",
        "is obligated to", "is prohibited from"
    ]

    text_lower = text.lower()
    found = {}
    for word in obligations:
        count = text_lower.count(word)
        if count > 0:
            found[word] = count

    # Sort by frequency
    return dict(sorted(found.items(),
                key=lambda x: x[1], reverse=True))


# ---- RUN ON OUR CONTRACT ----
print("🔍 Extracting entities from contract...\n")

entities = extract_legal_entities(CONTRACT_TEXT)
obligations = find_obligation_words(CONTRACT_TEXT)

# Display entities
print("👥 LEGAL ENTITIES FOUND:")
print("=" * 50)

entity_labels = {
    "ORG":     "🏢 Organizations",
    "PERSON":  "👤 People",
    "DATE":    "📅 Dates",
    "MONEY":   "💰 Money",
    "GPE":     "📍 Locations",
    "LAW":     "⚖️  Laws",
    "PERCENT": "📊 Percentages",
    "CARDINAL": "🔢 Numbers"
}

for label, values in entities.items():
    friendly = entity_labels.get(label, label)
    print(f"\n{friendly}:")
    for v in values[:5]:  # show top 5 of each
        print(f"  • {v}")

print("\n\n⚖️  OBLIGATION WORDS FREQUENCY:")
print("=" * 50)
for word, count in obligations.items():
    bar = "█" * min(count, 20)
    print(f"  '{word}': {bar} ({count}x)")
```

```python
# CELL 5C — Extract Key Contract Information

def extract_contract_summary(text, entities):
    """
    Uses extracted entities to build a structured
    summary of the contract's key information.
    """

    print("📋 CONTRACT KEY INFORMATION")
    print("=" * 50)

    # Document type from filename
    print(f"\n📄 Document: {CONTRACT_NAME}")

    # Parties involved
    orgs = entities.get("ORG", [])
    if orgs:
        print(f"\n🏢 PARTIES INVOLVED ({len(orgs)} organizations):")
        for org in orgs[:5]:
            print(f"   • {org}")

    # Important dates
    dates = entities.get("DATE", [])
    if dates:
        print(f"\n📅 KEY DATES MENTIONED ({len(dates)} found):")
        for date in dates[:5]:
            print(f"   • {date}")

    # Money amounts
    money = entities.get("MONEY", [])
    if money:
        print(f"\n💰 FINANCIAL TERMS ({len(money)} found):")
        for m in money[:5]:
            print(f"   • {m}")

    # Locations/Jurisdiction
    locations = entities.get("GPE", [])
    if locations:
        print(f"\n📍 LOCATIONS/JURISDICTION:")
        for loc in locations[:5]:
            print(f"   • {loc}")

    # Laws mentioned
    laws = entities.get("LAW", [])
    if laws:
        print(f"\n⚖️  LAWS REFERENCED:")
        for law in laws[:5]:
            print(f"   • {law}")

    print("\n" + "=" * 50)
    print("✅ Entity extraction complete!")


extract_contract_summary(CONTRACT_TEXT, entities)
```

```python
# CELL 6A — Split contract into clauses

import re

def split_into_clauses(text):
    """
    Splits contract into individual clauses.
    Legal contracts use numbered sections like:
    1. Something
    2. Something else
    1.1 Sub clause
    """

    # Split by numbered patterns
    clauses = re.split(
        r'\n(?=\s*\d+[\.\)]\s+[A-Z])',
        text
    )

    # Clean each clause
    cleaned = []
    for clause in clauses:
        clause = clause.strip()
        # Keep only meaningful clauses (not too short)
        if len(clause) > 80:
            cleaned.append(clause)

    print(f"✅ Found {len(cleaned)} clauses in contract")
    print("\n📋 FIRST 3 CLAUSES PREVIEW:")
    print("=" * 50)
    for i, clause in enumerate(cleaned[:3]):
        print(f"\n--- CLAUSE {i+1} ---")
        print(clause[:200])
        print("...")

    return cleaned


# Run it
clauses = split_into_clauses(CONTRACT_TEXT)
```

```python
# CELL 6B — Risk Classifier using Groq (Updated)

from groq import Groq
import os
import time

client = Groq(api_key=os.environ["GROQ_API_KEY"])

def classify_clause_risk(clause_text):
    """
    Sends each clause to Groq LLM and gets:
    1. Risk level (Safe / Caution / High Risk)
    2. Risk type
    3. Plain English explanation
    """

    prompt = f"""You are a legal contract risk analyzer.
Analyze this contract clause and respond in EXACTLY this format:

RISK_LEVEL: [Safe / Caution / High Risk]
RISK_TYPE: [one short phrase describing the risk type]
EXPLANATION: [one plain English sentence explaining why]

Contract clause:
{clause_text[:600]}"""

    try:
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[{"role": "user", "content": prompt}]
        )

        result_text = response.choices[0].message.content.strip()

        # Parse response
        lines = result_text.split('\n')
        risk_level  = "Unknown"
        risk_type   = "Unknown"
        explanation = "Unable to parse"

        for line in lines:
            if line.startswith("RISK_LEVEL:"):
                risk_level = line.replace("RISK_LEVEL:", "").strip()
            elif line.startswith("RISK_TYPE:"):
                risk_type = line.replace("RISK_TYPE:", "").strip()
            elif line.startswith("EXPLANATION:"):
                explanation = line.replace("EXPLANATION:", "").strip()

        # Add emoji
        if "High Risk" in risk_level:
            emoji = "🔴"
        elif "Caution" in risk_level:
            emoji = "🟡"
        else:
            emoji = "🟢"

        return {
            "risk_level": f"{emoji} {risk_level}",
            "risk_type": risk_type,
            "explanation": explanation
        }

    except Exception as e:
        return {
            "risk_level": "⚠️ Error",
            "risk_type": "API Error",
            "explanation": str(e)
        }


# Test on first clause
print("🔍 Testing risk classifier...\n")
test = classify_clause_risk(clauses[0])
print(f"Risk Level  : {test['risk_level']}")
print(f"Risk Type   : {test['risk_type']}")
print(f"Explanation : {test['explanation']}")
print("\n✅ Classifier working!")
```

```python
# CELL 6C — Analyze ALL clauses

print("⚖️  ANALYZING ALL CLAUSES FOR RISK...")
print("=" * 60)

risk_results = []
high_risk_count = 0
caution_count   = 0
safe_count      = 0

for i, clause in enumerate(clauses[:10]):
    print(f"Analyzing clause {i+1}/{min(10, len(clauses))}...",
          end=" ")

    result = classify_clause_risk(clause)
    result["clause_text"]   = clause[:200]
    result["clause_number"] = i + 1
    risk_results.append(result)

    print(result['risk_level'])
    time.sleep(0.5)  # small delay

# Full report
print("\n\n📊 FULL RISK ANALYSIS REPORT:")
print("=" * 60)

for r in risk_results:
    print(f"\nClause {r['clause_number']}: {r['risk_level']}")
    print(f"  Type       : {r['risk_type']}")
    print(f"  Explanation: {r['explanation']}")
    print(f"  Text       : {r['clause_text'][:100]}...")

    if "High Risk" in r['risk_level']:
        high_risk_count += 1
    elif "Caution" in r['risk_level']:
        caution_count += 1
    else:
        safe_count += 1

print("\n" + "=" * 60)
print("📈 RISK SUMMARY:")
print(f"  🔴 High Risk : {high_risk_count}")
print(f"  🟡 Caution   : {caution_count}")
print(f"  🟢 Safe      : {safe_count}")
print(f"  📄 Total     : {len(risk_results)}")
print("=" * 60)
print("\n✅ Phase 4 Complete!")
```

```python
# CELL 8 — Plain Language Summarizer using Groq

def summarize_in_plain_english(text, doc_type="legal contract"):
    """
    Sends contract to Groq and gets back a
    plain English summary anyone can understand.
    """

    prompt = f"""You are a legal document simplifier.
Read this {doc_type} and explain it in plain simple English
that a non-lawyer can understand.

Structure your response EXACTLY like this:

WHAT IS THIS: (one sentence describing the document)
KEY PARTIES: (who is involved)
MAIN OBLIGATIONS: (what each party must do)
IMPORTANT DATES: (key deadlines and durations)
MONEY INVOLVED: (all financial terms)
WATCH OUT FOR: (3 things the reader should be careful about)
SIMPLE SUMMARY: (2-3 sentence plain English summary)

Legal document:
{text[:3000]}"""

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content


# Run summarizer
print("📝 Generating plain English summary...")
print("=" * 60)

summary = summarize_in_plain_english(CONTRACT_TEXT,
                                      "affiliate agreement")
print(summary)
print("=" * 60)
print("\n✅ Phase 5 Complete!")

# Save summary for report later
CONTRACT_SUMMARY = summary
```

```python
# CELL 9 — Setup Vector Database for Q&A

!pip install sentence-transformers chromadb -q

from sentence_transformers import SentenceTransformer
import chromadb
import os

print("Loading embedding model...")
embedder = SentenceTransformer('all-MiniLM-L6-v2')
print("✅ Embedding model loaded!")

# FIX: Use local Colab storage instead of Google Drive
LOCAL_DB_FOLDER = "/content/chromadb/"
os.makedirs(LOCAL_DB_FOLDER, exist_ok=True)

chroma_client = chromadb.PersistentClient(path=LOCAL_DB_FOLDER)
print("✅ ChromaDB initialized!")

# Create fresh collection
try:
    chroma_client.delete_collection("legal_doc")
except:
    pass

collection = chroma_client.create_collection("legal_doc")

def chunk_text(text, chunk_size=500, overlap=50):
    """Splits text into overlapping chunks"""
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i+chunk_size])
        if len(chunk) > 100:
            chunks.append(chunk)
    return chunks

def index_document(text):
    """Converts document into embeddings and stores in ChromaDB"""
    chunks = chunk_text(text)
    print(f"📦 Indexing {len(chunks)} chunks...")

    embeddings = embedder.encode(chunks).tolist()

    collection.add(
        documents=chunks,
        embeddings=embeddings,
        ids=[f"chunk_{i}" for i in range(len(chunks))]
    )
    print(f"✅ Document indexed successfully!")
    return len(chunks)

# Index our contract
index_document(CONTRACT_TEXT)
```

```python
# CELL 10 — Q&A Function + Test (Fixed)

def answer_question(question):
    # Step 1: Embed question
    question_embedding = embedder.encode([question]).tolist()

    # Step 2: Find relevant chunks
    results = collection.query(
        query_embeddings=question_embedding,
        n_results=3
    )
    relevant_chunks = "\n\n".join(results['documents'][0])

    # Step 3: Ask Groq with context
    prompt = f"""You are a legal document assistant.
Answer the question based ONLY on the document excerpts below.
If the answer is not in the document say:
This information is not mentioned in the document.
Be concise and clear. Avoid legal jargon.

DOCUMENT EXCERPTS:
{relevant_chunks}

QUESTION: {question}

ANSWER:"""

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content.strip()


# Test questions
print("💬 DOCUMENT Q&A TEST")
print("=" * 60)

test_questions = [
    "Who are the parties involved in this agreement?",
    "What happens if I put restricted content on my website?",
    "How does Chase pay the affiliate?",
    "Can Chase terminate this agreement?"
]

for question in test_questions:
    print(f"\n❓ {question}")
    answer = answer_question(question)
    print(f"💡 {answer}")
    print("-" * 40)

print("\n✅ Phase 6 Complete!")
```

```python
# CELL 11 — PDF Report Generator (Complete Fixed Version)

!pip install fpdf2 -q

from fpdf import FPDF
import datetime
import os

def clean_str(text):
    return str(text).encode('latin-1', errors='replace').decode('latin-1')

def generate_report(doc_name, entities, risk_results, summary, output_path):
    pdf = FPDF()
    pdf.add_page()

    # HEADER
    pdf.set_fill_color(30, 30, 80)
    pdf.set_text_color(255, 255, 255)
    pdf.set_font("Helvetica", "B", 18)
    pdf.cell(0, 15, "Legal Document Analysis Report", new_x="LMARGIN", new_y="NEXT", align="C", fill=True)

    pdf.set_text_color(0, 0, 0)
    pdf.set_font("Helvetica", "", 10)
    pdf.cell(0, 7, clean_str("Document: " + doc_name[:60]), new_x="LMARGIN", new_y="NEXT")
    pdf.cell(0, 7, "Generated: " + datetime.datetime.now().strftime('%d %b %Y %H:%M'), new_x="LMARGIN", new_y="NEXT")
    pdf.ln(4)

    # RISK OVERVIEW
    high = sum(1 for r in risk_results if "High" in r.get('risk_level', ''))
    caut = sum(1 for r in risk_results if "Caution" in r.get('risk_level', ''))
    safe = sum(1 for r in risk_results if "Safe" in r.get('risk_level', ''))

    pdf.set_fill_color(240, 240, 240)
    pdf.set_font("Helvetica", "B", 12)
    pdf.cell(0, 10, "Risk Overview", new_x="LMARGIN", new_y="NEXT", fill=True)
    pdf.set_font("Helvetica", "", 11)
    pdf.cell(60, 8, "High Risk: " + str(high), new_x="RIGHT", new_y="TOP")
    pdf.cell(60, 8, "Caution: " + str(caut), new_x="RIGHT", new_y="TOP")
    pdf.cell(60, 8, "Safe: " + str(safe), new_x="LMARGIN", new_y="NEXT")
    pdf.ln(4)

    # ENTITIES
    pdf.set_fill_color(200, 220, 255)
    pdf.set_font("Helvetica", "B", 12)
    pdf.cell(0, 10, "Key Entities Identified", new_x="LMARGIN", new_y="NEXT", fill=True)
    pdf.set_font("Helvetica", "", 10)

    for label, values in entities.items():
        if values:
            pdf.set_font("Helvetica", "B", 10)
            pdf.cell(0, 6, str(label) + ":", new_x="LMARGIN", new_y="NEXT")
            pdf.set_font("Helvetica", "", 10)
            for v in list(values)[:3]:
                pdf.cell(10)
                pdf.cell(0, 6, clean_str("- " + str(v)), new_x="LMARGIN", new_y="NEXT")
    pdf.ln(4)

    # RISK CLAUSES
    pdf.set_fill_color(255, 200, 200)
    pdf.set_font("Helvetica", "B", 12)
    pdf.cell(0, 10, "Clause Risk Analysis", new_x="LMARGIN", new_y="NEXT", fill=True)

    for i, r in enumerate(risk_results[:10]):
        level = r.get('risk_level', 'Unknown')
        rtype = r.get('risk_type', 'Unknown')
        line  = "Clause " + str(i+1) + ": " + str(level) + " | " + str(rtype)
        pdf.set_font("Helvetica", "B", 9)
        pdf.cell(0, 6, clean_str(line), new_x="LMARGIN", new_y="NEXT")

        explanation = r.get('explanation', '')
        if explanation:
            pdf.set_font("Helvetica", "", 8)
            pdf.multi_cell(0, 5, clean_str(explanation[:200]))
        pdf.ln(2)
    pdf.ln(4)

    # SUMMARY
    pdf.set_fill_color(200, 255, 200)
    pdf.set_font("Helvetica", "B", 12)
    pdf.cell(0, 10, "Plain Language Summary", new_x="LMARGIN", new_y="NEXT", fill=True)
    pdf.set_font("Helvetica", "", 10)
    pdf.multi_cell(0, 6, clean_str(summary[:2000]))

    # SAVE
    pdf.output(output_path)
    print("Report saved to: " + output_path)


# RUN
report_path = "/content/legal_analysis_report.pdf"

generate_report(
    doc_name     = "contract.pdf",
    entities     = entities,
    risk_results = risk_results,
    summary      = summary,
    output_path  = report_path
)

# VERIFY AND DOWNLOAD
if os.path.exists(report_path):
    size = os.path.getsize(report_path)
    print("File confirmed - " + str(size) + " bytes")
    from google.colab import files
    files.download(report_path)
    print("Download started - check your Downloads folder!")
else:
    print("File not created - paste any error above")
```
